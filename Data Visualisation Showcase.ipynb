#### 4.1.1. Bar chart distribution of customer ages <a id="part4.1.1"></a>
Our customers range from ages 18 to 50 inclusive, with many of them being youths aged 35 and below, as seen from the approximately right-skewed distribution. Notably, the most common ages are 23 years old (18 customers), 25 years old (25 customers), and 26 years old (18 customers). I also observed that there are no customers aged 29 or 49.

```python
# Initialise the x and y values.
x_values = df['Age'].unique()
y_values = df.groupby('Age')['Age'].count()

# Create the bar chart. Set the title.
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(x_values, y_values, align='center')
ax.set_title('Bar chart distribution of customer ages', fontsize=20)

# Set the labels and limits.
ax.set_ylabel('Number of customers with age', fontsize=15)
ax.set_xlabel('Ages', fontsize=15)
ax.set_ylim(bottom=0, top=27)
ax.set_xlim(left=16, right=52)

# Annotate each bar with the count.
for x,y in zip(x_values ,y_values):
    label = '{:d}'.format(y)
    plt.annotate(label, 
                 (x,y), 
                 textcoords='offset points', 
                 xytext=(0,10), 
                 ha='center') 

# Display the figure.
fig.tight_layout()
plt.show()
```

#### 4.1.2. Swarm plot distribution of products bought by age <a id="part4.1.2"></a>
The 3 products are primarily purchased by younger customers. Specifically, product TM195 is most commonly bought by customers aged 26, while products TM498 and TM798 are predominantly purchased by those aged 25. Additionally, purchases from older customers are minimal, with only 8 transactions recorded from individuals aged 45 and above, based on manual inspection.

```python
# Set the figure size and create the swarm plot.
plt.figure(figsize=(10,6))
sns.swarmplot(x='Product', y='Age', data=df, hue='Product')

# Create the title, change the size of axes labels, and display figure.
plt.title('Swarm plot distribution of products bought by age', fontsize=20)
plt.xlabel('Product', fontsize=15)
plt.ylabel('Ages', fontsize=15)
plt.show()
```

#### 4.1.3. Horizontal bar chart of number of customers based on gender <a id="part4.1.3"></a>
Altogether, there are 104 male and 76 female customers, or in percentages, 57.78% of customers are male while 42.22% of customers are female. This analysis serves as a foundation for other analysis and visualisations.

```python
# Initialise the x and y values.
gender_count = df.groupby('Gender')['Gender'].count()
genders = gender_count.index

# Create the horizontal bar chart. Set the limits and title.
plt.figure(figsize=(10,6))
plt.barh(genders, gender_count)
plt.xlim(left=0,right=120)
plt.title('Number of customers based on gender', fontsize=15)

# Append the data value (count) at the end of each bar.
for x,y in zip(genders ,gender_count):
    label = y
    plt.annotate(label, 
                 (y,x), 
                 textcoords='offset points',
                 xytext=(25,0),
                 ha='center')

# Display the figure.
plt.show()
```

#### 4.1.4. Stacked bar chart of product sales by gender <a id="part4.1.4"></a>
In terms of best-sellers, TM195 is the most popular choice between both genders (40 units each). On the other hand, worst sellers differ for both genders, where it is TM498 for males (31 units), and it is TM798 for females (7 units). 
Without looking at gender, the best-selling product is TM195 (80 units), followed by TM498 (60 units), and lastly TM798 (40 units). 

```python
# Initialise the x and y values.
x_labels = df['Product'].unique()
male_purchase = df[df['Gender']=='Male'].groupby('Product')['Gender'].count().tolist()
female_purchase = df[df['Gender']=='Female'].groupby('Product')['Gender'].count().tolist()

# Set the width. Create the bar graph. Also add the title and legend.
width = 0.4
fig, ax = plt.subplots(figsize=(10,6))
bar1 = ax.bar(x_labels, male_purchase, width, label='Male', color = 'cyan')
bar2 = ax.bar(x_labels, female_purchase, width, 
              bottom=male_purchase, label='Female', color = 'pink')
plt.title('Stacked bar chart of sales by gender', fontsize=15)
plt.legend(loc=(1.02,0))

# Create a function to display data labels/annotations for each bar.
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        # Calculate the center of the bar
        center_x = bar.get_x() + bar.get_width() / 2
        center_y = bar.get_y() + height / 2
        ax.annotate('{}'.format(height),
                    xy=(center_x, center_y),
                    ha='center', va='center', color='black')
        
# Use function to display data labels for both bars.
autolabel(bar1)
autolabel(bar2)

# Display the figure itself.
fig.tight_layout()
plt.show()
```

#### 4.1.5. Pie chart of product sales among males <a id="part4.1.5"></a>
Even though [part 4.1.4](#part4.1.4) states that TM498 is the worst seller for males, the sales of products are somewhat evenly distributed. The deviation from expected sales (33.3% per product) is minimal, with a maximum deviation of 3.5% in total products sold.

```python
# Initialise the list of values.
x_labels = df['Product'].unique()
male_purchase = df[df['Gender']=='Male'].groupby('Product')['Gender'].count().tolist()

# Create the pie chart, create a title, and display it.
plt.pie(male_purchase, explode=(0.1,0.1,0.1),
        labels=x_labels, autopct='%1.1f%%', shadow=True)
plt.title('Breakdown of product sales among males', fontsize=20)
plt.show()
```

#### 4.1.6. Pie chart of product sales among females <a id="part4.1.6"></a>
On the other hand, the sales of products for females are uneven. It is clear that product TM195 is leading in terms of sales for females (52.6% of sales), and that product TM798 is selling poorly (9.2% of sales) by looking at the proportion to total sales. 

```python
# Initialise the list of values.
x_labels = df['Product'].unique()
female_purchase = df[df['Gender']=='Female'].groupby('Product')['Gender'].count().tolist()

# Create the pie chart, create a title, and display it.
plt.pie(female_purchase, explode=(0.1,0.1,0.1),
        labels=x_labels, autopct='%1.1f%%', shadow=True)
plt.title('Breakdown of product sales among females', fontsize=20)
plt.show()
```

#### 4.1.7. Box plot of product usage across all customers <a id="part4.1.7"></a>
Around 50% of customers expect to use our products 3 to 4 times a week, based on the interquartile range. Several outliers are observed, where some customers stated they expect to use our products 6 or 7 times a week. #### 4.3.2. Chi-square testing of education level and product purchased <a id="part4.3.2"></a>
To test if the number of years of formal education is associated with the product purchased, I will conduct the Chi-square test of independence, a statistical hypothesis testing method that tests the association and independence of 2 variables.

A 2-way table will be created, by grouping the years of formal education into 3 categories – Low, Medium, High.

```python
# Display a table, grouped by education level and product bought.
df_crosstab = pd.crosstab(df['Education'], df['Product'])
print(df_crosstab)
```

From the cross-table, I manually calculate the observed counts (*O*) of each cell. The 2-way table as follows.

Education | TM195 | TM498 | TM798 | TOTAL
--- | --- | --- | --- | ---
Low (14≤) | 35 | 26 | 2 | 63
Medium (15-17) | 43 | 32 | 15 | 90
High (≥18) | 2 | 2 | 23 | 27
TOTAL | 80 | 60 | 40 | 180

Next, the expected counts (*E*) is calculated, by the following formula:

*Expected count for a cell (E) = $\frac{RC}{G}$, where R = row total, C = column total, G = grand total*

The expected counts are in the brackets of each cell.

Education | TM195 | TM498 | TM798 | TOTAL
--- | --- | --- | --- | ---
Low (14≤) | 35 (28) | 26 (21) | 2 (14) | 63
Medium (15-17) | 43 (40) | 32 (30) | 15 (20) | 90
High (≥18) | 2 (12) | 2 (9) | 23 (6) | 27
TOTAL | 80 | 60 | 40 | 180

The null (${H}_{0}$) and alternate (${H}_{1}$) hypothesis are as follows:

- ${H}_{0}:$ Years of formal education and product purchased are independent.
- ${H}_{1}:$ Years of formal education and product purchased are associated.

For this test, I will calculate at a 5% level of significance, which refers to the probability that a null hypothesis (${H}_{0}$) is falsely rejected.

To determine rejection of ${H}_{0}$ , I will calculate the chi-square test statistic, which is given by the following formula:

$X^{2} = \sum \frac{(O-E)^{2}}{E}$

I will also calculate the critical value. If the critical value is more than the test statistic, then there is insufficient evidence to reject ${H}_{0}$ and they are independent. However, if the critical value is less than the test statistic, than there is sufficient evidence to reject ${H}_{0}$ and they are associated.

```python
# Store the observed and expected counts in a list. Initialise a test statistic variable.
observed = [35, 26, 2, 43, 32, 15, 2, 2, 23]
expected = [28, 21, 14, 40, 30, 20, 12, 9, 6]
test_statistic = 0

# Calculate the test statistic based on the formula.
for i in range(len(observed)):
    test_statistic += ((observed[i] - expected[i]) ** 2) / expected[i]

# The critical value is calculated at 9.49 based on a Chi-square distribution table.
critical_val = 9.49

# Display the statistics (degree of freedom, significance level, critical value, test statistic).
print('Degree of freedom: 4')
print('Significance level: 5%')
print(f'Critical value: {critical_val}')
print(f'Chi-test statistic: {test_statistic}\n')

# Display conclusion.
if critical_val > test_statistic:
    print('Critical value is larger than test statistic. Null hypothesis is not rejected.')
    print('Level of education and product purchased are independent.')
else:
    print('Critical value is smaller than test statistic. Null hypothesis is rejected.')
    print('Alternate hypothesis is accepted. Level of education and product purchased are associated.')
```

Based on the conclusion, the level of education and the product purchased are associated, and they have a significant relationship.

As such, a grouped bar chart was selected as the visualisation, as it easily allows for side-by-side comparison between genders and products, thus pointing out best sellers for each gender.

#### 5.1.3. Which age group(s) tend to use our products more than other age group(s)? <a id="part5.1.3"></a>
##### Analysis charts: [part 4.1.7](#part4.1.7), [part 4.1.8](#part4.1.8)
This visualisation answers question 1.3 of the primary exploratory questions. Some core findings and insights from the analysis in [part 4](#part4) include:
- The interquartile range of usage is between 3 to 4 times a week. ([part 4.1.7](#part4.1.7))
- The correlation coefficient between age and usage is very close to 0, indicating no linear relationship between both variables. ([part 4.1.8](#part4.1.8))

Given the findings, a simple bar chart of the average weekly usage by age group is chosen, as it is the simplest way for stakeholders (sales managers) to compare usage. Simultaneously, age group is chosen over age as age is a numerical measure, and there is too little data per age to deduce a reliable weekly usage representative of the entire age population.

For this visualisation, age groups with an average weekly usage lower than the overall weekly usage are red, and those higher than the overall average are green. As such, to answer the question, the **age group of 26 to 35 years generally uses our products more than other age groups**, as it is the only age group with an average usage higher than the overall average. 

```python
# Initialise a dictionary, age_dictionary, which will store usage values by defined age groups.
age_dictionary = {'18-25':[], '26-35':[], '35-43':[], '43-50':[]}

# Loop through each record, appending usage value into the corresponding age group's list in age_dictionary.
for age, usage in zip(df['Age'], df['Usage']):
    if age <= 25:
        age_dictionary['18-25'].append(usage)
    elif age <= 35:
        age_dictionary['26-35'].append(usage)
    elif age <= 43:
        age_dictionary['35-43'].append(usage)
    else:
        age_dictionary['43-50'].append(usage)

# Calculate the average usage for each age group.
for key, values in age_dictionary.items():
    grp_avg_usage = 0
    for usages in age_dictionary[key]:
        grp_avg_usage += usages
    grp_avg_usage = grp_avg_usage / len(age_dictionary[key])
    age_dictionary[key] = grp_avg_usage

# Calculate the overall average usage.
overall_avg_usage = df['Usage'].mean()

# Create the figure.
plt.figure(figsize=(10,6))

# Create the bar chart. 
for grp, usage in age_dictionary.items():
    # If an age group's average usage is below overall average, it is red in color, otherwise it is green.
    if usage < overall_avg_usage:
        plt.bar(grp, usage, color='red')
    else:
        plt.bar(grp, usage, color='green')
        
    # Label each bar with the average usage of that age group.
    label = "{:.2f}".format(usage)
    plt.annotate(label,
                 (grp,usage),
                 fontsize=12,
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

# Set the limit of the y-axis, to focus on numerical differences.
plt.ylim(bottom=3, top=4)

# Create the title and axes labels, adjusting the font sizes simultaneously.
plt.title('Average weekly usages by Age Groups', fontsize=20)
plt.ylabel('Average Usage', fontsize=15)
plt.xlabel('Age Group', fontsize=15)

# Display the figure.
plt.show()
```

### 6.1. Customer demopraphics and product preferences dashboard <a id="part6.1"></a>
This dashboard aims to answer all primary exploratory questions listed under [topic 1](#part2.2.1) (i.e. questions 1.1, 1.2, 1.3, 1.4), which discusses customer demographics and product preferences. It provides detailed information about:
- Product sale by Age (answering question 1.1);
- Product sale by Gender (answering question 1.2);
- Each Age Group’s average weekly Usage (answering question 1.3); and
- Customers’ Age and Gender distribution (answering question 1.4)

Potential answers for each primary exploratory question are listed under each visualisation of [part 5.1](#part5.1).

[Back to top](#top)

```python
# Create a figure to be used as a dashboard.
fig = plt.figure(figsize=(10,8), constrained_layout=True)

# Use GridSpec to determine the dimensions of the dashboard.
gs = GridSpec(nrows=2, ncols=2, figure=fig)

# -----------------------------------------------------------------------------

# 1st subplot - violin plot of product sale by age.
ax1 = fig.add_subplot(gs[0,0])

# Create and plot the violin and swarm plots.
product_palette = {'TM195':'cyan', 'TM498':'Orange', 'TM798':'#90EE90'}
sns.violinplot(x='Product', y='Age', data=df, inner=None, hue='Product', 
               palette=product_palette, ax=ax1)
sns.swarmplot(x='Product', y='Age', data=df, color='black', alpha=0.6, size=3.5, ax=ax1)

# Set the title. Also adjust the font size of title and axes labels.
ax1.set_title('Product sale by Age', fontsize=15)
ax1.set_xlabel('Product', fontsize=12)
ax1.set_ylabel('Age', fontsize=12)

# -----------------------------------------------------------------------------

# 2nd subplot - grouped bar chart of product sale by gender.
ax2 = fig.add_subplot(gs[1,0])

# Define the x labels and the values for purchases of products by gender (in 2 separate lists).
x_labels = df['Product'].unique()
male_purchase = df[df['Gender']=='Male'].groupby('Product')['Gender'].count().tolist()
female_purchase = df[df['Gender']=='Female'].groupby('Product')['Gender'].count().tolist()

# Arrange x_labels in fixed intervals. Define width for later.
x = np.arange(len(x_labels))
width = 0.4

# Create the grouped bar charts.
bar1 = ax2.bar(x - width/2, male_purchase, width, color='blue', label='Male') 
bar2 = ax2.bar(x + width/2, female_purchase, width, color='magenta', label='Female')

# Set the title and axes labels, adjusting their font sizes.
ax2.set_title('\nProduct sale by Gender', fontsize=15)
ax2.set_xlabel('Product', fontsize=12)
ax2.set_ylabel('Sales count', fontsize=12)

# Set the number of ticks for the x_labels. Set the y-limits to fit dashbaord.
ax2.set_xticks(x)
ax2.set_xticklabels(x_labels)
ax2.set_ylim(bottom=0, top=45)

# Call for the legend for easier identification of which color a gender is.
ax2.legend(loc=[0,0], title='Gender')

# Create a function, autolabel, to label each bar with the data count and percentage of total sales.
def autolabel(bars):
    for each in bars:
        height = each.get_height()
        ax2.annotate('{0} \n({1:.2f}%)'.format(height, height/1.8),
                    fontsize=9,
                    xy=(each.get_x() + each.get_width() / 2, height),
                    xytext=(0, 2),  # 2 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Call the function on bars of both genders.
autolabel(bar1)
autolabel(bar2)

# -----------------------------------------------------------------------------

# 3rd subplot - bar chart of average weekly usage by age groups.
ax3 = fig.add_subplot(gs[0,1])

# Initialise a dictionary, age_dictionary, which will store usage values by defined age groups.
age_dictionary = {'18-25':[], '26-35':[], '35-43':[], '43-50':[]}

# Loop through each record, appending usage value into the corresponding age group's list in age_dictionary.
for age, usage in zip(df['Age'], df['Usage']):
    if age <= 25:
        age_dictionary['18-25'].append(usage)
    elif age <= 35:
        age_dictionary['26-35'].append(usage)
    elif age <= 43:
        age_dictionary['35-43'].append(usage)
    else:
        age_dictionary['43-50'].append(usage)

# Calculate the average usage for each age group.
for key, values in age_dictionary.items():
    grp_avg_usage = 0
    for usages in age_dictionary[key]:
        grp_avg_usage += usages
    grp_avg_usage = grp_avg_usage / len(age_dictionary[key])
    age_dictionary[key] = grp_avg_usage

# Calculate the overall average usage.
overall_avg_usage = df['Usage'].mean()

# Create the bar chart. 
for grp, usage in age_dictionary.items():
    # If an age group's average usage is below overall average, it is red in color, otherwise it is green.
    if usage < overall_avg_usage:
        ax3.bar(grp, usage, color='red')
    else:
        ax3.bar(grp, usage, color='green')

    # Label each bar with the average usage of that age group.
    label = "{:.2f}".format(usage)
    ax3.annotate(label,
                 (grp,usage),
                 fontsize=9,
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

# Set the limit of the y-axis, to focus on numerical differences.
ax3.set_ylim(bottom=3, top=4)

# Create the title and axes labels, adjusting the font sizes simultaneously.
ax3.set_title('Average weekly usages by Age Groups', fontsize=15)
ax3.set_ylabel('Average Usage', fontsize=12)
ax3.set_xlabel('Age Group', fontsize=12)

# -----------------------------------------------------------------------------

# 4th subplot - Stacked histogram with KDE plot of age and gender distribution.
ax4 = fig.add_subplot(gs[1,1])

# Create the histogram with KDE plot.
gender_palette = {'Male':'Blue', 'Female':'Magenta'}
sns.histplot(data=df, x='Age', hue='Gender', kde=True, multiple='stack', palette=gender_palette, 
             bins=33, ax=ax4)

# Create the title and axes labels, adjusting their font sizes simultaneously.
ax4.set_title('\nCustomers\' Age and Gender distribution', fontsize=15)
ax4.set_xlabel('Ages', fontsize=12)
ax4.set_ylabel('Customer count', fontsize=12)

# -----------------------------------------------------------------------------

# Create the dashboard title, adjust its font size, and display the dashboard figure.
fig.suptitle('Customer demographics and product preferences\n', fontsize=20)
plt.show()
```
